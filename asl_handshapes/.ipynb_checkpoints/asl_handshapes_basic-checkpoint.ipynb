```python
data_dir = '/mnt/data/ppaudyal/handshape_color'
train_dir = os.path.join(data_dir, 'D')
validation_dir = os.path.join(data_dir,'A')
```

```python
#build a basic model 
from keras import layers, models, optimizers
def build_model():
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', 
                            input_shape=(150, 150, 3)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3,3), activation='relu'))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D(2,2))
    model.add(layers.Conv2D(128, (3,3), activation='relu'))
    model.add(layers.MaxPooling2D(2,2))
    model.add(layers.Flatten())
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(24, activation='softmax'))
    return model 
```

```python
ImageDataGenerator = keras.preprocessing.image.ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
validation_generator = test_datagen.flow_from_directory(validation_dir, target_size=(150, 150), batch_size=20, class_mode='categorical')
```

```python
train_folders = ['E', 'B', 'E', 'D']
histories = []
model = build_model()
model.compile(optimizer=optimizers.RMSprop(lr=1e-4), metrics=['acc'], loss='binary_crossentropy')
for t_f in train_folders:
    train_dir = os.path.join(data_dir, t_f)
    train_generator = train_datagen.flow_from_directory(train_dir, target_size=(150, 150), batch_size=20, class_mode='categorical') 
    histories.append(model.fit_generator(train_generator, epochs=15, steps_per_epoch=20, validation_data=validation_generator))
model.save('asl_handshape_scratch_4_06050219.h5')
```

```python
import pickle
with open ('histories_firstABC_4.pkl', 'wb') as pklfile:
    pickle.dump(histories, pklfile)
```

```python
from keras.applications import VGG16
conv_base = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150,3))
```

```python
import os
import numpy as np 
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1./255)
batch_size=10

def extract_features(directory, sample_count):
    print(directory)
    features = np.zeros(shape=(sample_count, 4, 4, 512))
    labels = np.zeros(shape=(sample_count, 24))
    generator = datagen.flow_from_directory(
                directory, 
                target_size=(150, 150),
                batch_size=batch_size,
                class_mode='categorical')
    i=0
    for inputs_batch, labels_batch in generator:
        features_batch = conv_base.predict(inputs_batch)
        features[i*batch_size:(i+1)*batch_size] = features_batch
        labels[i*batch_size:(i+1)*batch_size] = labels_batch
        i += 1
        if i* batch_size >= sample_count:
            break
    return features, labels
```

```python
def build_model_convnet_featureuser():
    model = models.Sequential()
    model.add(layers.Dense(256, activation='relu', input_dim=4*4*512))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(24, activation='softmax'))
    
    model.compile(optimizer = optimizers.RMSprop(lr=2e-5), 
                 loss = 'binary_crossentropy', 
                 metrics = ['acc'])
    return model
```

```python
data_dir = '/mnt/data/ppaudyal/handshape_color'
train_folders = ['E', 'A', 'C', 'D']
validation_dir = os.path.join(data_dir, 'B')
validation_features, validation_labels = extract_features(validation_dir, 450)
validation_features = np.reshape(validation_features, (450, 4*4*512))
model_with_convbase = build_model_convnet_featureuser()
histories_convbases = [] 
for i in range(5):
    for t_f in train_folders:
        train_dir = os.path.join(data_dir, t_f)
        train_features, train_labels = extract_features(train_dir, 4000)
        train_features = np.reshape(train_features, (4000, 4*4*512))
        histories_convbases.append(model_with_convbase.fit(train_features, train_labels, epochs=30, batch_size=10, validation_data=(validation_features, validation_labels)))
model_with_convbase.save('first_a_b_c_model.h5')
```

```python
import pickle
with open ('histories_convbase_4.pkl', 'wb') as pklfile:
    pickle.dump(histories_convbases, pklfile)
```
